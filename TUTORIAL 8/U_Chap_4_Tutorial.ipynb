```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('titanic.csv')
titanic = sns.load_dataset("titanic")
```

```python
df.info()
```

```python
df.columns
```

```python
titanic.info()
```

```python
sns.lineplot(data=titanic, x="age", y="fare")
plt.show()
```

```python
sns.scatterplot(x="age", y="fare", data=titanic)
plt.show()
```

```python
sns.countplot(x="class", data=titanic)
plt.show()
```

```python
sns.barplot(x="class", y="fare", data=titanic, estimator=sum)
plt.show()
```

```python
sns.boxplot(x="class", y="age", data=titanic)
plt.show()
```

```python
sns.violinplot(x="class", y="age", data=titanic)
plt.show()
```

```python
sns.histplot(titanic["age"], kde=True)
plt.show()
```

```python
numeric_df = titanic.select_dtypes(include=["float64", "int64"])
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.pairplot(titanic.select_dtypes(include=["float64", "int64"]))
plt.show()
```

```python
sns.stripplot(x="class", y="age", data=titanic)
plt.show()
```

```python
sns.swarmplot(x="class", y="age", data=titanic)
plt.show()
```
